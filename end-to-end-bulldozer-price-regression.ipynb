#### split data in to train and validation

```python
df_valid = df_temp[df_temp.saleyear == 2012]
df_train = df_temp[df_temp.saleyear != 2012]
```

#### split data in to x and y

```python
x_train,y_train = df_train.drop("SalePrice",axis = 1),df_train.SalePrice
x_valid,y_valid = df_valid.drop("SalePrice",axis=1),df_valid.SalePrice
```

#### create evaluation function

```python
from sklearn.metrics import mean_squared_log_error,mean_absolute_error,r2_score
def rmsle(y_test,y_preda):
    return np.sqrt(mean_squared_log_error(y_test,y_preda))
def show_score(model):
    train_preda = model.predict(x_train)
    valid_preda = model.predict(x_valid)
    scores = {"Training MAE":mean_absolute_error(y_train,train_preda),
             "Valid MAE":mean_absolute_error(y_valid,valid_preda),
             "Training RMSLE":rmsle(y_train,train_preda),
             "Valid RMSLE":rmsle(y_valid,valid_preda),
             "Training R^2":r2_score(y_train,train_preda),
             "Valid R^2":r2_score(y_valid,valid_preda)}
    return scores
```

```python
model = RandomForestRegressor(n_jobs = -1,
                             random_state = 80,
                             max_samples = 10000)
```

```python
%%time
model.fit(x_train,y_train)
```

```python
show_score(model)
```

```python
%%time
from sklearn.model_selection import RandomizedSearchCV
rc_dict = {"n_estimators":np.arange(10,100,10),
          "max_depth":[None,3,5,10],
          "min_samples_split":np.arange(2,20,2),
          "min_samples_leaf":np.arange(1,20,2),
          "max_features":[0.5,1,"sqrt","auto"],
          "max_samples":[10000]}
rs_model = RandomizedSearchCV(RandomForestRegressor(n_jobs = -1,
                                                   random_state = 80),
                             param_distributions = rc_dict,
                             cv = 5,
                             n_iter = 2,
                             verbose = True)
rs_model.fit(x_train,y_train)
```

```python
# find best hyperparameter of model
rs_model.best_params_
```

```python
show_score(rs_model)
```

#### Train a model with the best hyperparametwr

```python
%%time
ideal_model = RandomForestRegressor(n_estimators = 10,
                                   min_samples_split = 16,
                                   min_samples_leaf = 13,
                                   max_features = 0.5,
                                   n_jobs = -1,
                                   max_samples = None)
ideal_model.fit(x_train,y_train)
```

### Preprocessing data(getting test dataset in to same formate of training datasets)

```python
def preprocess(df):
    df["saleyear"] = df.saledate.dt.year
    df["salemonth"] = df.saledate.dt.month
    df["saleday"] = df.saledate.dt.day
    df["saledayofweek"] = df.saledate.dt.dayofweek
    df["saledayofyear"] = df.saledate.dt.dayofyear
    
    df.drop("saledate" , axis = 1, inplace = True)
    
    #fill numeric row with median
    for label,content in df.items():
        if pd.api.types.is_numeric_dtype(content):
            if pd.isnull(content).sum():
                df[label + "_is_missing"] = pd.isnull(content)
                df[label] = content.fillna(content.median())
                
            #fill categorical missing data and turn categorical data in to numbers
        if not pd.api.types.is_numeric_dtype(content):
            df[label + "_is_missing"] = pd.isnull(content)
            df[label] = pd.Categorical(content).codes+1
    return df
```

### Features Importance

```python
#function for feature importance

def plot_features_importance(columns,importances,n=20):
    df = (pd.DataFrame({"features":columns,
                       "features_importance":importances})
         .sort_values("features_importance",ascending = False)
         .reset_index(index = True))
    fig,ax = plt.subplots()
    ax.barh(df["features"][:n],df["features_importance"][:20])
    ax.set_ylabel("Features")
    ax.set_xlabel("Features_Importance")
```
